We will explore the following questions:
<br>Do students in more popular majors make more money?
<br>Do students that majored in subjects that were majority female make more money?
<br>Is there any link between the number of full-time employees and median salary?

```python
#Median vs Sample size
recent_grads.plot(x='Sample_size',y='Median',kind='scatter',title='Sample_size vs Median')
```

```python
#Unemployment_rate vs Sample size
recent_grads.plot(x='Sample_size',y='Unemployment_rate',kind='scatter',title='Sample_size vs Median')
```

```python
#Full_time vs Median
recent_grads.plot(y='Full_time',x='Median',kind='scatter',title='Full_time vs Median')
```

```python
#ShareWomen and Unemployment_rate
recent_grads.plot(y='Unemployment_rate',x='ShareWomen',kind='scatter',title='ShareWomen and Unemployment_rate')
```

```python
#Men and Median
recent_grads.plot(y='Median',x='Men',kind='scatter',title='Men and Median')
```

```python
#Women and Median
recent_grads.plot(y='Median',x='Women',kind='scatter',title='Women and Median')
```

```python
#Do students in more popular majors make more money?
recent_grads.plot(y='Median',x='Total',kind='scatter',title='Do students in more popular majors make more money?')
#it does not seem to be the case
```

```python
#Do students that majored in subjects that were majority female make more money?
recent_grads.plot(y='Median',x='ShareWomen',kind='scatter',title='Do students that majored in subjects that were majority female make more money?')
#there seems to be an inverse correlation between median pay and the share of women in a certain major
```

```python
#Is there any link between the number of full-time employees and median salary?
recent_grads.plot(y='Median',x='Full_time',kind='scatter',title='Is there any link between the number of full-time employees and median salary?')
```
